```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('uber_cleaned.csv')

# Basic info
print("Dataset shape:", df.shape)
print(df.info())

# Descriptive stats for Hour
print("\nHour Stats:")
print(df['Hour'].describe())

# Count rides by Hour
hour_counts = df['Hour'].value_counts().sort_index()
print("\nRides count by Hour:\n", hour_counts)

# Count rides by Weekday
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts = df['Weekday'].value_counts().reindex(weekday_order)
print("\nRides count by Weekday:\n", weekday_counts)

# Plot: Number of rides by Hour
plt.figure(figsize=(10,5))
sns.barplot(x=hour_counts.index, y=hour_counts.values, palette="viridis")
plt.title('Number of Rides by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.show()

# Plot: Number of rides by Weekday
plt.figure(figsize=(10,5))
sns.barplot(x=weekday_counts.index, y=weekday_counts.values, palette="magma")
plt.title('Number of Rides by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Number of Rides')
plt.show()

# Plot: Geographic distribution of pickups (scatter plot)
plt.figure(figsize=(8,8))
sns.scatterplot(x='Lon', y='Lat', data=df.sample(5000), alpha=0.3, s=10)
plt.title('Sample of Ride Pickup Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
